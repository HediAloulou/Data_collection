```python
import pandas as pd
```

```python
data = pd.read_csv("_sensor_data.csv")
```

```python
filtered_df = data[data['username'] == 'aymen']
```

```python
filtered_df.head()
```

```python
filtered_df=filtered_df.sort_values('timestamp')
```

```python
filtered_df.head(20)
```

```python
filtered_df=filtered_df.reset_index(drop=True)
```

```python
filtered_df.head(30)
```

```python
filtered_df['sensor'].value_counts()
```

```python
filtered_df.drop_duplicates(subset=['timestamp', 'sensor'], inplace=True)
```

```python
filtered_df['sensor'].value_counts()
```

```python
filtered_df.head(20)
```

```python
linear_acceleration_sensors = []
rotation_sensors = []
gyroscope_sensors = []

# Iterate through dataframe
for index, row in filtered_df.iterrows():
    sensor_type = row['sensor']
    if sensor_type == 'Linear Acceleration':
        linear_acceleration_sensors.append(row.tolist())  # Append entire row as list
    elif sensor_type == 'Rotation Vector':
        rotation_sensors.append(row.tolist())  # Append entire row as list
    elif sensor_type == 'Gyroscope':
        gyroscope_sensors.append(row.tolist())  
```

```python
gyroscope_sensors
```

```python
linear_acceleration_sensors
```

```python
linear_acceleration_sensors
```

```python
min_len = min(len(linear_acceleration_sensors), len(rotation_sensors), len(gyroscope_sensors))

# Create a combined list of lists where each row contains information from all three lists
combined_sensors = []
for i in range(min_len):
    print(linear_acceleration_sensors)
    combined_row = linear_acceleration_sensors[i] + rotation_sensors[i] + gyroscope_sensors[i]
    combined_sensors.append(combined_row)
```

```python
combined_sensors
```

```python
# Define columns for the new dataframe
new_columns = (filtered_df.columns.tolist() * 3)  # Repeat the original columns 3 times

# Create a new dataframe from the combined list of lists
combined_df = pd.DataFrame(combined_sensors, columns=new_columns)
```

```python
combined_df
```

```python
combined_df = combined_df.rename(columns={"x": "x_linear_acceleration", "y": "y_linear_acceleration","z": "z_linear_acceleration"})
```

```python
combined_df
```

```python
combined_df.to_csv('combined_sensors.csv', index=False)

print("Dataframe has been saved to combined_sensors.csv")
```
