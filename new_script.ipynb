```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv("_sensor_data.csv")
```

```python
data = data.sort_values('timestamp')
```

```python
data = data.reset_index(drop=True)
```

```python
data
```

```python
data = data.dropna(axis=0)
```

```python
data
```

```python
data['sensor'].value_counts()
```

```python
linear_acceleration_sensors = []
rotation_sensors = []
gyroscope_sensors = []

# Iterate through dataframe
for index, row in data.iterrows():
    sensor_type = row['sensor']
    if sensor_type == 'Linear Acceleration':
        linear_acceleration_sensors.append(row.tolist())  # Append entire row as list
    elif sensor_type == 'Rotation Vector':
        rotation_sensors.append(row.tolist())  # Append entire row as list
    elif sensor_type == 'Gyroscope':
        gyroscope_sensors.append(row.tolist())  
```

```python
min_len = min(len(linear_acceleration_sensors), len(rotation_sensors), len(gyroscope_sensors))

# Create a combined list of lists where each row contains information from all three lists
combined_sensors = []
for i in range(min_len):
    combined_row = linear_acceleration_sensors[i] + rotation_sensors[i] + gyroscope_sensors[i]
    combined_sensors.append(combined_row)
```

```python
combined_df = pd.DataFrame(combined_sensors, columns=new_columns)
```
